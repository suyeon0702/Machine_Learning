```python
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

train_set, test_set = split_train_test(nba2, 0.2)

#ì˜ˆì¸¡ ë³€ìˆ˜ì™€ ë ˆì´ë¸”ì˜ ë¶„ë¦¬
nba_data = train_set.drop("Points", axis=1)
nba_labels = train_set["Points"].copy()

#nba_dataì—ì„œ í…ìŠ¤íŠ¸ íŠ¹ì„±ì„ ì œì™¸í•œ ë³µì‚¬ë³¸
nba_num = nba_data.drop("Player", axis=1)
nba_num = nba_num.drop("Pos", axis=1)
nba_num = nba_num.drop("Team", axis=1)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return self

num_pipeline = Pipeline([
    ('std_scaler', StandardScaler()),
])

nba_num_tr = num_pipeline.fit(nba_num)
nba_num_tr = num_pipeline.transform(nba_num)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(nba_num)
cat_attribs = ["Player", "Pos", "Team"]

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])

nba_prepared = full_pipeline.fit(nba_data)
nba_prepared = full_pipeline.transform(nba_data)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

tree_reg = DecisionTreeRegressor()
tree_reg.fit(nba_prepared, nba_labels)

scores = cross_val_score(tree_reg, nba_prepared, nba_labels,
                        scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
tree_rmse_scores
```

```python
def display_scores(scores):
    print("ì ìˆ˜:", scores)
    print("í‰ê· :", scores.mean())
    print("í‘œì¤€íŽ¸ì°¨:", scores.std())
    
display_scores(tree_rmse_scores)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=10, random_state=42)
forest_reg.fit(nba_prepared, nba_labels)
```

```python
from sklearn.metrics import mean_squared_error

nba_predictions = forest_reg.predict(nba_prepared)
forest_mse = mean_squared_error(nba_labels, nba_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
from sklearn.model_selection import cross_val_score

forest_scores = cross_val_score(forest_reg, nba_prepared, nba_labels,
                                scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```
